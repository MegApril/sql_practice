```python
import sqlite3
import csv
```

Initial creation adding employee values

```python
data = [
    (101, 'Carol', 'Analyst', 70000),
    (102, 'Mike', 'Engineer', 75000),
    (103, 'Joseph', 'Sales Manager', 80000)
]
```

```python
def add_data_to_database():
  try:
    conn = sqlite3.connect('company_demo_data.db')
    cursor = conn.cursor()
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS employees(
        ID PRIMARY KEY,
        names TEXT,
        position TEXT,
        salary REAL
    )
    """
    )
    cursor.executemany("""
    INSERT INTO employees VALUES(?,?,?,?)
    """, data
    )
    conn.commit()
    print("Information added succesfully")
  except Exception as e:
    print(e)
  finally:
    conn.close()


add_data_to_database()
```

Viewing employee data

```python
conn = sqlite3.connect('company_demo_data.db')
cursor = conn.cursor()
cursor.execute("""
SELECT *
FROM employees
""")
# gets all data that is added
# cursor.fetchall()

# fetchmany() takes an argument of size to view only a certain number of rows
results = cursor.fetchmany(size=2)

for row in results:
  print(row)
```

Creation of company safety table using csv file

```python
def add_data_to_database(input_data):
  try:
    conn = sqlite3.connect('company_demo_data.db')
    cursor = conn.cursor()
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS safety(
        ID PRIMARY KEY,
        names TEXT,
        department TEXT,
        incident_date DATE,
        incident_type TEXT,
        days_lost INT,
        injury_severity TEXT
    )
    """
    )
    with open(input_data, 'r') as csv_file:
      csv_reader = csv.reader(csv_file)
      next(csv_reader)
      for row in csv_reader:
        cursor.execute("INSERT INTO safety VALUES(?,?,?,?,?,?,?)", row)
    conn.commit()
    print("Information added succesfully")
  except Exception as e:
    print(e)
  finally:
    conn.close()

data = "/content/drive/MyDrive/Colab Notebooks/SQL_Essentials_For_Data_Analysis_Datasets/company_safety.csv"
add_data_to_database(data)
```

```python
conn = sqlite3.connect('company_demo_data.db')
cursor = conn.cursor()
cursor.execute("""
SELECT *
FROM safety
""")
# gets all data that is added
results = cursor.fetchall()

# fetchmany() takes an argument of size to view only a certain number of rows
# results = cursor.fetchmany(size=2)

for row in results:
  print(row)
```

```python
import pandas as pd
import sqlite3
```

```python
def add_data_to_database(input_data):
  try:
    df = pd.read_csv(input_data)
    conn = sqlite3.connect('company_demo_data.db')
    df.to_sql('safety', con=conn, index=False)
    print("Information added successfully")
    conn.close()
  except Exception as e:
    print(e)


data = "/content/drive/MyDrive/Colab Notebooks/SQL_Essentials_For_Data_Analysis_Datasets/company_safety.csv"
add_data_to_database(data)
```
